```python
# ì‹ë‹¹ ì •ë³´ ë¦¬ìŠ¤íŠ¸ë¥¼ ë¶ˆëŸ¬ì˜¤ëŠ” í•¨ìˆ˜

def set_RT_List():
    sql_RT = "select * from RT_List"   # ì‹ë‹¹ì •ë³´ í…Œì´ë¸”ì— ìˆëŠ” ì •ë³´ë¥¼ ë¶ˆëŸ¬ì˜¤ëŠ” sqlë¬¸
    cursor.execute(sql_RT)   # ìœ„ sqlë¬¸ì„ ì‹¤í–‰ì‹œí‚¤ëŠ” executeí•¨ìˆ˜
    RT_List = cursor.fetchall()  # fetchallë¡œ ì‹¤í–‰ëœ sqlë¬¸ì˜ ì •ë³´ë¥¼ RT_Listì— ë‹´ëŠ”ë‹¤.
    return RT_List


# ì‹ë‹¹ ë¦¬ë·° ë¦¬ìŠ¤íŠ¸ë¥¼ ë¶ˆëŸ¬ì˜¤ëŠ” í•¨ìˆ˜

def set_RV_List():  # ì‹ë‹¹ ë¦¬ë·° í…Œì´ë¸”ì— ìˆëŠ” ì •ë³´ë¥¼ ë¶ˆëŸ¬ì˜¤ëŠ” sqlë¬¸
    sql_RV = "select * from RV_List"  
    cursor.execute(sql_RV)
    RV_List = cursor.fetchall()
    return RV_List


# ê°€ê²Œì´ë¦„ìœ¼ë¡œ í‰ê· ì ìˆ˜ êµ¬í•˜ê¸°

def avg_RV_List(RT_name):  # a ëŠ” ì‹ë‹¹ì´ë¦„
    RV_L = set_RV_List() # RV_Lì— ì‹ë‹¹ë¦¬ë·° ë¦¬ìŠ¤íŠ¸ë¥¼ ë‹´ëŠ” ì‘ì—…
    List = []  # ë¹ˆ ë¦¬ìŠ¤íŠ¸ ê³µê°„ì„ ìƒì„±
    
    for i in range(len(RV_L)): # ì‹ë‹¹ë¦¬ë·° ë¦¬ìŠ¤íŠ¸ì˜ ê¸¸ì´ ë§Œí¼ ë°˜ë³µ
        
        if RV_L[i][0]== RT_name:  # ì‹ë‹¹ë¦¬ë·° ë¦¬ìŠ¤íŠ¸ì˜ ì´ë¦„ê³¼ ì…ë ¥ë°›ì€ ì´ë¦„ì´ ê°™ë‹¤ë©´
            List.append(RV_L[i])  # í•´ë‹¹ ì‹ë‹¹ë¦¬ë·°ë¥¼ ë¹ˆ ë¦¬ìŠ¤íŠ¸ ê³µê°„ì— ë‹´ëŠ”ë‹¤.
    
    tot = 0  # í‰ì ì˜ í•©ì˜ ì´ˆê¸°ê°’ì„ 0ìœ¼ë¡œ ë‘ .
    
    for j in List:   # ìœ„ì— ìƒì„±ëœ ë¦¬ìŠ¤íŠ¸ë¥¼ í•œì¤„ì”© ë°˜ë³µí•˜ëŠ” ë°˜ë³µë¬¸
        tot += j[4]   # ì´ì ì€ í•´ë‹¹ ì‹ë‹¹ í‰ì ì˜ í•©
    avg = tot/len(List)   # í•´ë‹¹ ì‹ë‹¹ í‰ì ì˜ í•©ì„ ìœ„ì— ë¦¬ìŠ¤íŠ¸ì— ë‹´ê¸´ ì‹ë‹¹ ë¦¬ë·°ì˜ ê°¯ìˆ˜ë¡œ ë‚˜ëˆˆë‹¤.
            
    return avg
```
